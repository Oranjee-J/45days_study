### 1️⃣ 魔法方法：字符串输出

定义一个 `Person` 类，包含属性 `name` 和 `age`。

要求：

- 实现 `__str__` 方法

- 当 `print(Person(...))` 时输出：

  ```
  Name: xxx, Age: xxx
  ```

````python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def __str__(self):
        return (f'Name: {self.name},Age:{self.age}')

print(Person('Alice',18))
````

### 2️⃣ 魔法方法：长度定义

定义一个 `Team` 类，内部有成员列表 `members`。

要求：

- 实现 `__len__`
- 使得 `len(team)` 返回成员数量

````python
class Team:
    def __init__(self, members):
        self.members = members

    def __len__(self):
        return len(self.members)

m = ['a','b','c','d']
print(len(Team(m))) 
````

### 3️⃣ 魔法方法：索引访问

定义一个 `BookShelf` 类，内部有书名列表 `books`

要求：

- 实现 `__getitem__`
- 支持 `shelf[0]` 返回第一本书

````python
class BookShelf:
    def __init__(self, books):
        self.books = books
    
    def __getitem__(self,index):
        return self.books[index]

m = ['asdasd','qwdwqdqwdq','sdcswa','sdvdfdfg']

print((BookShelf(m))[0])
````

### 4️⃣ 魔法方法：加法运算

定义一个 `Vector` 类，包含 `x, y`

要求：

- 实现 `__add__`
- 使 `v1 + v2` 返回新的 `Vector`

````python
class Vector:
    def __init__(self, x,y):
        self.x = x
        self.y = y

    def __add__(self,other):
        return(self.x+other.x,self.y+other.y)

v1 = Vector(1,2)

v2 = Vector(3,4)

print(v1+v2)
````

### 5️⃣ 闭包：基础计数器

写一个函数 `counter()`：

要求：

- 返回一个内部函数 `add()`
- 每调用一次 `add()`，计数 +1
- 返回当前计数

````python
def counter():
    counter = 0

    def inner(func):
        
        nonlocal counter
        func()
        counter+=1


        return counter
    

    return inner


def type():
    print('haha')


print(counter(type))
````
