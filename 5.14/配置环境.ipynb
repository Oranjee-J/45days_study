# 配置conda环境

### 1.下载anaconda
*注意事项：如果本地已有python安装，和anaconda不冲突，直接下载即可。*

anaconda下载版本注意选择windows-x86_64的版本，这里是清华镜像源的地址，这个地址下载相对于官网来说更快和流畅。
```http
https://mirrors.tuna.tsinghua.edu.cn/anaconda/archive/
```
    
双击安装包，然后一直选择继续；在自己使用的电脑上，安装面对用户选择建议选择All Users，这样切换用户无需重新安装。我的conda安装地址为
```
C:\ProgramData\anaconda3
```

安装完成后，先不要运行，关闭运行程序，然后打开```设置-系统信息-高级系统设置-环境变量```来配置环境变量。
选择```path```进入，点击```新建```，输入
```
D:\anaconda3
D:\anaconda3\Scripts
D:\anaconda3\Library\bin
D:\anaconda3\Library\mingw-w64\bin
```
新建这四个环境变量，保存。

然后以管理员身份打开```anaconda promt```来验证是否完成安装。
```bash
conda --version
```
如果显示了正确的conda版本，那么即安装完成

### 2.正确配置anaconda环境
*我的用户名是中文，如果直接在默认路径下创建虚拟环境，会报错，无法运行conda init，无法激活虚拟环境。所以在安装完成之后，要将环境目录和包缓存目录修改到不含中文字符的路径下。*

打开anaconda powershell 
   ```bash
      conda config --add envs_dirs C:\conda_envs
      conda config --add pkgs_dirs C:\conda_pkg
   ```

创建的时候最好指定路径创建，这样不容易被保存到默认路径里。这里以安装python3.8版本的虚拟环境为例子：
   ```bash
      conda create -p C:\conda_envs\env1-py38 python=3.8 -y
   ```

激活时也建议加上物理路径，更加严谨
   ```bash
      conda activate C:\conda_envs\env1-py38
   ```


##### 创建完成后需要验证python版本和python的位置，还要查看所有环境
激活之后进入了虚拟环境，按照以下步骤输入powershell命令：
   ```bash
      python --version
   ```
如果正确地安装的话，python版本会显示为 ```3.8.x```
接下来查看python位置
   ```bash
      where.exe python
   ```
如果成功激活了虚拟环境，显示如下:   ``` C:\conda_envs\env1-py38\python.exe ```

接下来deactivate虚拟环境，退出到base环境里，查看环境列表，如果除了base还显示了我们创建的虚拟环境，那就是创建成功了。
   ```bash
      conda env list
   ```

删除环境时也加上物理路径
   ```bash
      conda remove -p C:\conda_envs\myenv --all
   ```

### 3. 在vscode里配置虚拟环境
在右下角python版本的边上可以选择虚拟环境，如果此时已经装好了虚拟环境，就会显示

``` Python 3.8.20 (env1-py38) C:\conda_envs\env1-py38\python.exe```

接下来新建一个python文件，运行如下代码来验证是不是成功切换到虚拟环境里了：

```python
import sys
print(sys.executable)
```

此时输出的内容为
```
C:\Users\小橘\AppData\Local\Programs\Python\Python314\python.exe
```

这证明并没有成功切换到虚拟环境中。

此时需要在vscode的终端里激活虚拟环境，然后才可使用。

在这一步我的终端报错，无法激活环境，这是因为我的系统默认不可执行.ps1脚本。执行以下命令：

```bash
Set-ExecutionPolicy -Scope CurrentUser RemoteSigned
```

如果成功执行，关闭所有终端，新建vscode终端，然后激活虚拟环境，在虚拟环境下运行此代码即可。

结果如果是```C:\conda_envs\env1-py38\python.exe```，那么就代表vscode能成功使用虚拟环境了。
