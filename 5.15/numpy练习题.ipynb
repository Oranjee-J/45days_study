## 11. 数组运算

创建两个 1~5 的数组，做加减乘除。

```python
a = np.arange(1,6)
b = np.arange(1,6)
print(np.add(a,b))      # a+b
print(np.subtract(a,b))     # a-b
print(np.multiply(a,b))     # a*b
print(np.divide(a,b))       # a/b
```

## 14.平均值

求一个随机数组的平均值。


```python

a = np.random.randint(0,10,(4,4))
print(a)
print(np.max(a,axis = 0))
print(np.average(a))
```

```python
a = np.random.randint(0,10,(4,4))

print(a)
print(np.max(a))
print(np.max(a,axis = 0))
print(np.max(a,axis = 1))
print(np.min(a))
print(np.average(a,axis=1))
```

## 17.转置

对一个矩阵进行转置。


```python
a = np.random.randint(1,10,(2,3,4))
print(a)
print(a.shape)
print(a.T)
print(a.T.shape)
```

```python
a = np.random.randint(1,10,(2,3))
print(a)
b = np.left_shift(a,1)
print(b)
c = np.right_shift(a,1)
print(c)
```

```python
a = np.random.randint(1,10,(2,3,4))
print(a)
print(a.shape)
b = np.transpose(a,axes = (1,0,2))      # b[i][j][k] = a[j][i][k]
print(b)
print(b.shape)
```

## 19.argmax

找出数组中最大值的索引。

```python
a = np.random.randint(1,20,(6,6))
print(a)
index = np.argwhere(a==6)
print(index)
index2 = np.unravel_index(np.argmax(a), a.shape)

print(index2)
m = np.argmax(a)
print(m)
```

## 20.综合题（稍微难一点）

给定数组：

[3, 7, 2, 9, 12, 5, 18, 1]

要求：

找出所有大于平均值的数
并按从大到小排序

```python
import numpy as np
a = np.array([3,7,2,9,18,5,12,1])
m = (a>np.average(a))
b = a[m]
print(np.sort(b))
```
