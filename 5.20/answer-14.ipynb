```python
import numpy as np

A = np.random.randn(3,3)
print(A)
B = A.T
print(B)
C = A @ B
print(C)

D = C.T
print(D)

def duichen(C, D):
    flag = 0
    for i in range(0,3):
        for j in range(0,3):
            if C[i][j]!=D[i][j] :
                flag = 1
                break
    
    return flag

result = duichen(C, D)
if result == 0:
    print('对称')
else:
    print('不对称')
```
